# Task 5. Future Sales Prediction 

## Machine learning model that can predict the future sales of a product using linear regression model.

```python
# Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Importing the dataset and displacing the first five lines of the dataset

df = pd.read_csv("future sales.csv")
print(df.head())
```

```python
# Shape of the data

df.shape
```

```python
# Statistical values

df.describe()
```

```python
# checking for the null values

df.isnull().sum()
```

```python
# Displacing the value count of each attribute

df.value_counts()
```

```python
# Plotting

plt.scatter(df.Sales,df.TV)
plt.xlabel('Sales ')
plt.ylabel('TV')
plt.title('Sales vs TV')
```

```python
# Plotting

plt.scatter(df.Sales,df.Newspaper)
plt.xlabel('Sales ')
plt.ylabel('Newspaper')
plt.title('Sales vs Newspaper')
```

```python
# Plotting

plt.scatter(df.Sales,df.Radio)
plt.xlabel('Sales ')
plt.ylabel('Radio')
plt.title('Sales vs Radio')
```

```python
# Finding the correlation and plotting

correlation = df.corr()
print(correlation["Sales"].sort_values(ascending=False))

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# independent and dependent variables
x = np.array(df[["Sales"]])
y = np.array(df["Sales"])
```

```python
# Splitting the dataset into train and test

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

## Model Building

```python
model = LinearRegression()
```

```python
#train the model
model.fit(x_train, y_train)
```

```python
# test the model
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
#Error
E=y_test-y_pred
E
```

```python
# Accuracy

from sklearn.metrics import r2_score
print("accuracy:",r2_score(y_pred,y_test)*100)
```
